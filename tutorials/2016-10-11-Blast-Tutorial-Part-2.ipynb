Need a query and database for BLAST.

1. use `makeblastdb` to create the database that we need
2. `in` the input file for the database
3. `dbtype` specifies the type of file, `prot`
4. `-out` puts the output in a specified directory

```python
!/Applications/ncbi-blast-2.5.0+/bin/makeblastdb \
-in /Users/yaaminivenkataraman/Documents/School/Year1/FISH-546/yaaminiv-fish546-2016/data/uniprot_sprot.fasta \
-dbtype prot \
-out /Users/yaaminivenkataraman/Documents/School/Year1/FISH-546/yaaminiv-fish546-2016/data/uniprot_sprot
```

1. specifies the program we want to use, `blastx`
2. `-query` provides the file we want to blast
3. `-db` specifies database created in the previous step
4. `-outfmt` type of file output. 6 is tabular

```python
!/Applications/ncbi-blast-2.5.0+/bin/blastx \
-query /Users/yaaminivenkataraman/Documents/School/Year1/FISH-546/yaaminiv-fish546-2016/data/Oly-transcriptome-2k.fa \
-db /Users/yaaminivenkataraman/Documents/School/Year1/FISH-546/yaaminiv-fish546-2016/data/uniprot_sprot \
-outfmt 6
```

`blastx` compares a nucleotide sequence (what we have) and allows us to compare it to protiens
